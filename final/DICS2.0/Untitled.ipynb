```python
import openai
import json

# 设置 OpenAI API 密钥
openai.api_key = "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"

query = '我對象最近生氣的時候會亂砸東西 我該怎麼辦才好'

# 定义函数以使用不同的模型生成回答
def get_response_from_model(model_name, prompt):
    response = openai.ChatCompletion.create(
        model=model_name,
        messages=[
            {"role": "assistant", "content": prompt},
            {"role": "user", "content": query}
        ],
        temperature=0.7,
        max_tokens=128,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0.6,
        stop=["\n"]
    )
    return response

# 定义合併策略函数
def merge_with_weights(responses, weights):
    prompt_for_gpt4 = "請問在128字以內將以下四個句子配合上權重綜合起來句子會變成如何？\n"
    for resp, weight in zip(responses, weights):
        sentence = resp['choices'][0]['message']['content']
        prompt_for_gpt4 += f"句子：{sentence}，權重：{weight}\n"
    
    merged_response = get_response_from_model("gpt-4", prompt_for_gpt4)
    return merged_response['choices'][0]['message']['content']

# 调用函数获取回答
models = [
    "ft:gpt-3.5-turbo-0613:pmu:recipe-ner:8EzN7CPl",
    "ft:gpt-3.5-turbo-0613:pmu:recipe-ner:8EzTh1dF",
    "ft:gpt-3.5-turbo-0613:pmu:recipe-ner:8Ezb91Yq",
    "ft:gpt-3.5-turbo-0613:pmu:recipe-ner:8EzVnZvJ"
]

prompt_description = "你是一個溫柔且很有耐心的女生，支持對方大部分的選擇，但是如果對方的選擇是有關傷害自己或者傷害他人的話，你會阻止他，時常給予他鼓勵，且是他的摯友"
responses = [get_response_from_model(model, prompt_description) for model in models]

# 输出格式化的回答
print("\n題目:", query)
print("="*40)
response_prefixes = ["d型回答", "i型回答", "c型回答", "s型回答"]
for prefix, response in zip(response_prefixes, responses):
    response_content = response['choices'][0]['message']['content']
    print(f"\n{prefix}:\n{response_content}")
    print("-"*40)
    
weights = [0.4,0.1, 0.4, 0.1]
merged_response = merge_with_weights(responses, weights)

print("\n綜合回答(使用的權重):", weights)
print("-"*40)
print(merged_response)
```
